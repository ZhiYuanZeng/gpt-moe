```python
import matplotlib.pyplot as plt
import numpy as np
import csv

data = []
labels = []

with open('data.csv') as f:
    reader = csv.reader(f)
    
    # label
    labels = next(reader) 
        
    for row in reader:
        data.append([float(x) for x in row])
print(labels)
print(data)
data = np.array(data).T
x = data[0]

fig, axes = plt.subplots(nrows=2, ncols=2)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

baseline = [0.3305, 0.6434, 0.2949, 0.2811]
for i in range(len(data)-1):
    ax = axes[i//2, i%2]

    y = data[i+1]
    label = labels[i+1]
    ax.plot(x, y, label=label)
    ax.axhline(baseline[i], color='r', linestyle='--', label='baseline')
    
    ax.set_xlabel('rank of expert selection')
    ax.legend()
plt.show()
```
